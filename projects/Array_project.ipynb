```python
import numpy as np
import matplotlib.pyplot as plt

# Sample weather data (Days vs. Temperature, Humidity, Weather Conditions)
days = np.array([1, 2, 3, 4, 5, 6, 7])  # Days of the week
temperature = np.array([22, 24, 19, 21, 23, 25, 22])  # Temperature in C
humidity = np.array([60, 65, 55, 70, 63, 59, 62])  # Humidity in %
weather_conditions = ['Sunny', 'Cloudy', 'Rainy', 'Sunny', 'Windy', 'Cloudy', 'Sunny']  # Weather conditions

# Create a figure with two subplots: Line Plot and Pie Chart
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Line Plot for Temperature and Humidity over Time
axs[0].plot(days, temperature, marker='o', label='Temperature (C)', color='orange')
axs[0].plot(days, humidity, marker='o', label='Humidity (%)', color='blue')
axs[0].set_title('Temperature and Humidity Over Time')
axs[0].set_xlabel('Day')
axs[0].set_ylabel('Value')
axs[0].legend()

# Pie Chart for Weather Conditions Distribution
weather_conditions_count = {condition: weather_conditions.count(condition) for condition in set(weather_conditions)}
axs[1].pie(weather_conditions_count.values(), labels=weather_conditions_count.keys(), autopct='%1.1f%%', startangle=140, colors=['#FF9999','#66B3FF','#99FF99','#FFCC99'])
axs[1].set_title('Weather Conditions Distribution')

# Adjust layout to prevent overlap
plt.title("Weather Data Analysis")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample weather data (Days vs. Temperature, Humidity, Weather Conditions)
days = np.array([1, 2, 3, 4, 5, 6, 7])  # Days of the week
temperature = np.array([22, 24, 19, 21, 23, 25, 22])  # Temperature in C
humidity = np.array([60, 65, 55, 70, 63, 59, 62])  # Humidity in %
weather_conditions = ['Sunny', 'Cloudy', 'Rainy', 'Sunny', 'Windy', 'Cloudy', 'Sunny']  # Weather conditions

# Create a figure with two subplots: Line Plot and Pie Chart
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Add a main title for the entire figure
fig.suptitle('Weekly Weather Analysis', fontsize=16)

# Line Plot for Temperature and Humidity over Time
axs[0].plot(days, temperature, marker='o', label='Temperature (C)', color='orange')
axs[0].plot(days, humidity, marker='o', label='Humidity (%)', color='blue')
axs[0].set_title('Temperature and Humidity Over Time')
axs[0].set_xlabel('Day')
axs[0].set_ylabel('Value')
axs[0].legend()

# Pie Chart for Weather Conditions Distribution
weather_conditions_count = {condition: weather_conditions.count(condition) for condition in set(weather_conditions)}
axs[1].pie(weather_conditions_count.values(), labels=weather_conditions_count.keys(), autopct='%1.1f%%', startangle=140, colors=['#FF9999','#66B3FF','#99FF99','#FFCC99'])
axs[1].set_title('Weather Conditions Distribution')

# Adjust layout to prevent overlap
plt.tight_layout()  # Adjust to make room for the main title
plt.show()
```
